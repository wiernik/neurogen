```python
from pathlib import Path
import pandas as pd
import numpy as np
from ChildProject.projects import ChildProject
from ChildProject.annotations import AnnotationManager

DATA_PATH = Path('/home/engaclew/neurogen')
columns = ['recording_filename', '5s_CTC', '1s_CTC', 'voc_chi', 'voc_dur_chi']

# Read measures
aclew_measures = pd.read_csv(DATA_PATH / 'aclew_measures_chunks.csv').fillna(0)
lena_measures = pd.read_csv(DATA_PATH / 'lena_measures_chunks.csv').fillna(0)
human_measures = pd.read_csv(DATA_PATH / 'human_measures_chunks.csv').fillna(0)

# Read metadata
children = pd.read_csv(DATA_PATH / 'data/L3_HIPAA_LENA_cleaned/metadata/children.csv')
recordings = pd.read_csv(DATA_PATH / 'data/L3_HIPAA_LENA_cleaned/metadata/recordings.csv')
recordings_data = recordings.merge(children, on='child_id')[['group_id', 'recording_filename']]
aclew_measures = aclew_measures.merge(recordings_data, how='left', on='recording_filename')
lena_measures = lena_measures.merge(recordings_data, how='left', on='recording_filename')
human_measures = human_measures.merge(recordings_data, how='left', on='recording_filename')

def group_data(data):
    if 'can_voc_CHI' in data.columns and 'non_can_voc_CHI' in data.columns:
        data['CVC'] = data['can_voc_CHI'] + data['non_can_voc_CHI']
    data = data.groupby('recording_filename').agg({
        '5s_CTC': np.sum,
        '1s_CTC': np.sum,
        'voc_dur_chi': np.sum,
        'voc_chi': np.sum,
        'wc_adu': np.sum,
        'CVC': np.sum,
        'group_id': 'first',
        'child_id': 'first'
    }).reset_index()
    return data

aclew_measures_gp = group_data(aclew_measures)
lena_measures_gp = group_data(lena_measures)
human_measures_gp = group_data(human_measures)
```

```python
from scipy import stats

def compute_pearson(x, y):
    """ Compute Pearson correlation coefficient """
    print(len(x), len(y))
    r, p = stats.pearsonr(x, y)
    return r, p

def compute_mae(x, y):
    """ Compute mean error """
    mae_list = (x-y).values
    average_mae = np.mean(mae_list)
    std_mae = np.std(mae_list)
    return mae_list, average_mae, std_mae

def compute_mae(x, y):
    """ Compute mean error """
    mae_list = (x-y).values
    average_mae = np.mean(mae_list)
    std_mae = np.std(mae_list)
    return mae_list, average_mae, std_mae

def compute_mape(x, y):
    """ Compute mean error """
    errors = np.zeros_like(x, dtype=float)
    # Case 1: y != 0 - standard MAPE calculation
    mask_nonzero = (y != 0)
    errors[mask_nonzero] = abs(x[mask_nonzero] - y[mask_nonzero]) / y[mask_nonzero]
    # Case 2: y = 0 and x != 0 -> error = 1
    mask_zero_error = (y == 0) & (x != 0)
    errors[mask_zero_error] = 1
    # Case 3: y = 0 and x = 0 -> error = 0 is already handled by initialization
    
    average_mape = np.mean(errors)
    std_mape = np.std(errors)
    return errors, average_mape, std_mape


columns = ['5s_CTC', '1s_CTC', 'voc_dur_chi', 'voc_chi', 'wc_adu', 'CVC']
lena_scores, aclew_scores = {}, {}
for col in columns:
    if col == 'voc_dur_chi':
        lena_measures[col] /= 1000
        aclew_measures[col] /= 1000
        human_measures[col] /= 1000

    # Compute LENA
    lena_pearson_r, lena_pearson_p = compute_pearson(lena_measures[col], human_measures[col])
    lena_mae_list, lena_average_mae, lena_std_mae = compute_mae(lena_measures[col], human_measures[col])
    lena_mape_list, lena_average_mape, lena_std_mape = compute_mape(lena_measures[col], human_measures[col])

    # Compute ACLEW
    aclew_pearson_r, aclew_pearson_p = compute_pearson(aclew_measures[col], human_measures[col])
    aclew_mae_list, aclew_average_mae, aclew_std_mae = compute_mae(aclew_measures[col], human_measures[col])
    aclew_mape_list, aclew_average_mape, aclew_std_mape = compute_mae(aclew_measures[col], human_measures[col])

    lena_scores[col] = {
        'average_mae': lena_average_mae,
        'std_mae': lena_std_mae,
        'average_mape': lena_average_mape,
        'std_mape': lena_std_mape,
        'pearson_r': lena_pearson_r,
        'pearson_p': lena_pearson_p,
        'mae_list': lena_mae_list,
        'mape_list': lena_mape_list
        }
    aclew_scores[col] = {
        'average_mae': aclew_average_mae,
        'std_mae': aclew_std_mae,
        'average_mape': aclew_average_mape,
        'std_mape': aclew_std_mape,
        'pearson_r': aclew_pearson_r,
        'pearson_p': aclew_pearson_p,
        'mae_list': aclew_mae_list,
        'mape_list': aclew_mape_list,
        }
    
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
import pandas as pd

matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['font.serif'] = 'Arial'
matplotlib.rcParams.update({'font.size': 14, 'legend.handleheight':1, 'hatch.linewidth': 1.0,
                           'lines.markersize':10, 'lines.linewidth':2, 'errorbar.capsize': 5})

cm = 1/2.54
lena_color = '#6664b8'
aclew_color = '#E69F5C'
metric_name = {
    '5s_CTC': 'CTC (5s)',
    '1s_CTC': 'CTC (1s)',
    'voc_dur_chi': 'Duration of the key child vocalizations',
    'voc_chi': 'Number of key child vocalizations',
    'wc_adu': 'AWC',
    'CVC': 'CVC'
}

def plot_mae_and_pearson(lena_scores, aclew_scores, metric):
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16*cm, 12*cm))
    width = 0.35
    
    # Create DataFrame for violin plot
    lena_errors = pd.DataFrame({
        'Algorithm': ['LENA'] * len(lena_scores[metric]['mae_list']),
        'Error': lena_scores[metric]['mae_list']
    })
    aclew_errors = pd.DataFrame({
        'Algorithm': ['ACLEW'] * len(aclew_scores[metric]['mae_list']),
        'Error': aclew_scores[metric]['mae_list']
    })
    plot_data = pd.concat([lena_errors, aclew_errors], ignore_index=True)
    
    # MAE plot with violin plots
    sns.violinplot(data=plot_data, x='Algorithm', y='Error', 
                  ax=ax1, 
                  palette=[lena_color, aclew_color],
                  inner='box',  # Shows quartile and median
                  linewidth=2,
                  saturation=1)
    
    if metric == 'voc_dur_chi':
        ax1.set_ylabel('Error (s)')
    else:
        ax1.set_ylabel('Error (count)')
    ax1.grid(True, linestyle='--', alpha=0.7, zorder=0)
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    ax1.set_xlabel('')  # Remove x label
    ax1.set_xticks([])  # Remove x ticks
    
    # Add MAE values above violins
    ymin, ymax = ax1.get_ylim()
    shift = .07 * (ymax-ymin)
    lena_mae = lena_scores[metric]['average_mae']
    aclew_mae = aclew_scores[metric]['average_mae']
    
    ax1.text(0, ymax, r'$\mu = $' + f'{lena_mae:.1f}', 
         ha='center', va='bottom', color=lena_color, weight='bold')
    ax1.text(1, ymax, r'$\mu = $' + f'{aclew_mae:.1f}', 
             ha='center', va='bottom', color=aclew_color, weight='bold')
    
    # Pearson's R plot
    lena_r = lena_scores[metric]['pearson_r']
    aclew_r = aclew_scores[metric]['pearson_r']
    
    ax2.bar([0 - width/2], 
            [lena_r], 
            width,
            color=lena_color, 
            edgecolor='black', 
            zorder=2)
    
    ax2.bar([0 + width/2], 
            [aclew_r], 
            width,
            color=aclew_color, 
            edgecolor='black',
            zorder=2)
    
    ax2.set_ylabel("Pearson's R")
    ax2.set_xticks([])
    ax2.grid(True, linestyle='--', alpha=0.7, zorder=0)
    ax2.set_ylim(0, 1)
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    
    # Add value labels for Pearson's R
    ymin, ymax = ax2.get_ylim()
    shift = .07 * (ymax-ymin)
    ax2.text(-width/2, lena_r - shift, 
            f'{lena_r:.2f}', 
            ha='center', va='bottom', color='white', weight='bold', zorder=3)
    ax2.text(width/2, aclew_r - shift, 
            f'{aclew_r:.2f}', 
            ha='center', va='bottom', color='white', weight='bold', zorder=3)
    
    # Create legend
    handles = [plt.Rectangle((0,0),1,1, color=lena_color), 
              plt.Rectangle((0,0),1,1, color=aclew_color)]
    labels = ['LENA', 'ACLEW']
    fig.legend(handles, labels, ncol=2, bbox_to_anchor=(0.5, 1.05), loc='center')
    
    # Add common xlabel
    fig.text(0.5, -0.03, metric_name[metric], ha='center')
    
metrics = ['5s_CTC', '1s_CTC', 'voc_dur_chi', 'voc_chi', 'wc_adu', 'CVC']
for metric in metrics:
    plot_mae_and_pearson(lena_scores, aclew_scores, metric)
    plt.tight_layout()
    plt.savefig(f'/home/engaclew/neurogen/figures/corpus_lena_vtc_{metric}.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib 
from scipy import stats
import numpy as np

matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['font.serif'] = 'Arial'
matplotlib.rcParams.update({'font.size': 14, 'legend.handleheight':1, 'hatch.linewidth': 1.0,
                           'lines.markersize':6, 'lines.linewidth':2, 'errorbar.capsize': 5})
cm = 1/2.54
fig, ax = plt.subplots(1, 1, figsize=(16*cm, 12*cm))

# Define colors
lena_color = '#6664b8'  # keeping original purple
aclew_color = '#FF8C00'  # bright orange, high contrast with purple

columns = ['5s_CTC', '1s_CTC', 'voc_chi', 'voc_dur_chi']
group_ids = ['low_risk', 'angelman_syndrome', 'fragile_x_syndrome', 'down_syndrome', 'autism_sibling']
col_name = {'5s_CTC': 'CTC (5s)',
            '1s_CTC': 'CTC (1s)',
            'voc_chi': 'number of CHI voc.',
            'voc_duration_chi': 'duration of CHI voc.',
           }
col = '5s_CTC'
group_id = 'angelman_syndrome'

def add_regression_lines(lena_x, aclew_x, human_y_lena, human_y_aclew):
    # Get common range for both x and y
    min_val = min(lena_x.min(), aclew_x.min(), human_y_lena.min(), human_y_aclew.min())
    max_val = max(lena_x.max(), aclew_x.max(), human_y_lena.max(), human_y_aclew.max())
    
    # Create range for regression lines
    plot_range = np.linspace(min_val, max_val, 100)
    
    # Add y=x line first (so it's behind other elements)
    plt.plot(plot_range, plot_range, 'k--', alpha=1, label='y = x')
    
    # Function to calculate regression and confidence interval
    def calc_regression(x, y, color, label):
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
        y_pred = slope * plot_range + intercept
        
        # Calculate prediction interval
        y_hat = slope * x + intercept
        mse = np.sum((y - y_hat) ** 2) / (len(x) - 2)
        x_mean = np.mean(x)
        x_std = np.sum((x - x_mean) ** 2)
        t = stats.t.ppf(0.975, len(x) - 2)
        pi = t * np.sqrt(mse * (1 + 1/len(x) + (plot_range - x_mean)**2 / x_std))
        
        plt.plot(plot_range, y_pred, color=color, alpha=0.9, label=label)  # increased alpha
        plt.fill_between(plot_range, y_pred - pi, y_pred + pi, color=color, alpha=0.15)  # slightly increased alpha
    
    # Calculate regression for both datasets
    calc_regression(lena_x, human_y_lena, lena_color, label='LENA')
    calc_regression(aclew_x, human_y_aclew, aclew_color, label='ACLEW')
    
    # Set same limits for both axes
    dist = max_val - min_val
    min_val = min_val -.1*dist
    max_val = max_val +.1*dist
    plt.xlim(min_val, max_val)
    plt.ylim(min_val, max_val)
    plt.legend(loc='upper left')
    
sub_lena = lena_measures[lena_measures.group_id == group_id]
sub_human = human_measures[human_measures.group_id == group_id]
sub_aclew = aclew_measures[aclew_measures.group_id == group_id]

plt.scatter(sub_lena[col], sub_human[col], color=lena_color, alpha=.8)  # increased alpha
plt.scatter(sub_aclew[col], sub_human[col], color=aclew_color, alpha=.8, marker='x')  # increased alpha
add_regression_lines(sub_lena[col], sub_aclew[col], 
                    sub_human[col], sub_human[col])
plt.xlabel(f'Automatic {col_name[col]}')
plt.ylabel(f'Human {col_name[col]}')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.title('Human vs. automatic estimates of\n Conversational Turn Count (CTC) for the Angelman group', fontsize=14)
plt.savefig('/home/engaclew/neurogen/figures/angelman_lena_vtc_CTC.png', dpi=300)
plt.show()
```
